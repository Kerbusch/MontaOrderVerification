```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
%matplotlib inline

os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"
```

```python
# Load a model
model = YOLO('MontaOrderVerification/oot_model9/weights/best.pt')  # load a custom model
```

```python
folder_path = '../Datasets/Dataset_OOT_YOLOv8/Test/'
for files in os.listdir( folder_path):
    # Predict with the model
    results = model.predict(os.path.join(folder_path, files),device='cpu')

    res_plotted = results[0].plot()

    # Convert the image from BGR to RGB
    image = cv2.cvtColor(res_plotted, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```
